```python
from urllib.request import urlopen, Request
from bs4 import BeautifulSoup
import requests
import pandas as pd
import numpy as np

# The original scraping and DataFrame creation code
url = 'https://serebii.net/pokedex-sv/dragon.shtml'
html_contents = requests.get(url).text
soup = BeautifulSoup(html_contents, 'lxml')
# Extract the table and clean data as you did previously
dextable = soup.find("table", attrs={"class": "dextable"})
dextable_rows = dextable.find_all("td")
for td in soup.find_all():
    if len(td.get_text(strip=True)) == 0:
        td.extract()
            
headings = []
for td in soup.find_all("td"):
    headings.append(td.text.replace('\n', ' ').strip())

data = headings[9:]
data.remove('Pic')
data.remove('Base Stats')
data.remove('Type')
h = len(data)
l = int(h / 9)
df = pd.DataFrame(np.array(data).reshape(l, 9))
# Set headers
df.columns = df.iloc[0]
df = df[1:]
df
```
